```python
import matplotlib.pyplot as plt
plt.style.use('classic')
```

```python
%matplotlib inline
import numpy as np
```

```python
#預設圖表圖
x=np.linspace(0,10,1000)
fig,ax=plt.subplots()
ax.plot(x,np.sin(x),'-b',label='Sine')
ax.plot(x,np.cos(x),'--r',label='Cosine')
ax.axis('equal')
leg=ax.legend();
```

```python
#多種方式自訂圖例--1
x=np.linspace(0,10,1000)
fig,ax=plt.subplots()
ax.plot(x,np.sin(x),'-b',label='Sine')
ax.plot(x,np.cos(x),'--r',label='Cosine')
ax.axis('equal')
ax.legend(loc='upper left',frameon=False);#指定位置以及關閉周圍的格線
```

```python
#多種方式自訂圖例--2
x=np.linspace(0,10,1000)
fig,ax=plt.subplots()
ax.plot(x,np.sin(x),'-b',label='Sine')
ax.plot(x,np.cos(x),'--r',label='Cosine')
ax.axis('equal')
ax.legend(frameon=False,loc='lower center',ncol=2);#ncol參數指定圖例的欄位數
```

```python
#多種方式自訂圖例--3
x=np.linspace(0,10,1000)
fig,ax=plt.subplots()
ax.plot(x,np.sin(x),'-b',label='Sine')
ax.plot(x,np.cos(x),'--r',label='Cosine')
ax.axis('equal')
ax.legend(fancybox=True,framealpha=1,shadow=True,borderpad=1);#圓角矩型(fancybox),陰影(shadow),透明度(alpha值)
```

```python
#選取圖例所要使用的元素
#方法一
plt.subplot(2,1,1)
y=np.sin(x[:,np.newaxis]+np.pi*np.arange(0,2,0.5))
lines=plt.plot(x,y)
plt.legend(lines[:2],['first','second']);#lines[:2]表只要表達第1條以及第2條線即可
#方法二
plt.subplot(2,1,2)
plt.plot(x,y[:,0],label='first')
plt.plot(x,y[:,1],label='second')
plt.plot(x,y[:,2:])
plt.legend(framealpha=1,frameon=True);
```

```python
#在圖例中的資料點大小
import pandas as pd
cities=pd.read_csv('/content/drive/MyDrive/DataScientist/Practice/PythonDataScienceHandbook/data/california_cities.csv')
#取感興趣的data
lat,lon=cities['latd'],cities['longd']#取經度與緯度
population,area=cities['population_total'],cities['area_total_km2']
#使用大小以及顏色但沒有標籤,畫上資料點
plt.scatter(lon,lat,label=None,
      c=np.log10(population),cmap='viridis',
      s=area,linewidth=0,alpha=0.5)
plt.xlabel('longtitude')
plt.ylabel('latitude')
plt.colorbar(label='log$_{10}$(population)')
plt.clim(3,7);
#以下建立一個圖例:
#畫上一個有我們想要大小以及標籤的空串列
for area in [100,300,500]:
  plt.scatter([],[],c='k',alpha=0.3,s=area,
        label=str(area)+'km$^2$')
plt.legend(scatterpoints=1,frameon=False,
      labelspacing=1,title='City Area')
plt.title('California Cities:Area and Population');
```

```python
#多重圖例
fig,ax=plt.subplots()
lines=[]
styles=['-','--','-.',':']
x=np.linspace(0,10,1000)
#建立4種線條
for i in range(4):
  lines+=ax.plot(x,np.sin(x-i*np.pi/2),
          styles[i],color='black')
#指定第一個圖例的線條和標籤
ax.legend(lines[:2],['lineA','lineB'],
     loc='upper right',frameon=False)
#建立第2個圖例,然後手動加入artist
from matplotlib.legend import Legend
leg=Legend(ax,lines[2:],['lineC','lineD'],
     loc='lower right',frameon=False)
ax.add_artist(leg);
```
