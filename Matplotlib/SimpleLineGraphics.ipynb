```python
#簡單的線性圖形
#%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
#建立figure和axes
fig=plt.figure()
ax=plt.axes()
#函式繪出資料
plt.subplot(2,1,1)
x=np.linspace(0,10,1000)
plt.plot(x,np.sin(x))
#多個線條
plt.subplot(2,1,2)
y=np.linspace(0,10,1000)
plt.plot(y,np.sin(y))
plt.plot(y,np.cos(y));
```

```python
#調整圖表:顏色與樣式
#如果沒有指定就會拿一組顏色循環使用
plt.subplot(4,1,1)
plt.plot(x,np.sin(x-0),color='blue')
plt.plot(x,np.sin(x-1),color='g')
plt.plot(x,np.sin(x-2),color='0.75')
plt.plot(x,np.sin(x-3),color='#FFDD44')
plt.plot(x,np.sin(x-4),color=(1.0,0.2,0.3))
plt.plot(x,np.sin(x-5),color='chartreuse');
#linestyle設定線條樣式
plt.subplot(4,1,2)
plt.plot(x,x+0,linestyle='solid')
plt.plot(x,x+1,linestyle='dashed')
plt.plot(x,x+2,linestyle='dashdot')
plt.plot(x,x+3,linestyle='dotted');
#或是用編碼的方式
plt.subplot(4,1,3)
plt.plot(x,x+0,linestyle='-')
plt.plot(x,x+1,linestyle='--')
plt.plot(x,x+2,linestyle='-.')
plt.plot(x,x+3,linestyle=':');
#更加簡潔作法(顏色與樣式合併使用)
plt.subplot(4,1,4)
plt.plot(x,x+0,'-g')
plt.plot(x,x+1,'--c')
plt.plot(x,x+2,'-.k')
plt.plot(x,x+3,':r');
```

```python
#調整圖表:Axes範圍
#利用xlim與ylim
plt.subplot(5,1,1)
plt.plot(x,np.sin(x))
plt.xlim(-1,11)
plt.ylim(-1.5,1.5);
#axis反向顯示
plt.subplot(5,1,2)
plt.plot(x,np.sin(x))
plt.xlim(10,0)
plt.ylim(1.2,-1.2);
#一次設定
plt.subplot(5,1,3)
plt.plot(x,np.sin(x))
plt.axis([-1,11,-1.5,1.5]);
#axis使圖更加緊密
plt.subplot(5,1,4)
plt.plot(x,np.sin(x))
plt.axis('tight');
#axis使圖相同比例(x的單位等於y的單位)
plt.subplot(5,1,5)
plt.plot(x,np.sin(x))
plt.axis('equal');
```

```python
#為圖表加上標籤
plt.subplot(2,1,1)
plt.plot(x,np.sin(x))
plt.title("A Sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)");
#在圖列上標記線條類型(legend()+label)
plt.subplot(2,1,2)
plt.plot(x,np.sin(x),'-g',label='sin(x)')
plt.plot(y,np.cos(x),':b',label='cos(x)')
plt.axis("equal")
plt.legend();
```

```python
#ax.set()一次設定所有屬性
ax=plt.axes()
ax.plot(x,np.sin(x))
ax.set(xlim=(0,10),ylim=(-2,2),
       xlabel='x',ylabel='sin(x)',
       title='A Simple Plot');
```
