```python
#MATLAB型式的介面
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
x=np.linspace(0,10,100)
fig=plt.figure()
#建立2個panel中的第一個
plt.subplot(2,1,1)#(建立2列1行的格子,將該圖放置於第一個)
plt.plot(x,np.sin(x))
#建立2個panel中的第二個
plt.subplot(2,1,2)#(建立2列1行的格子,將該圖放置於第二個)
plt.plot(x,np.cos(x));
```

```python
#物件導向式介面
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
x=np.linspace(0,10,100)
#建立一個圖表的格子
#ax為2個軸物件的陣列
fig,ax=plt.subplots(2)
#在適當物件上呼叫plot()方法
ax[0].plot(x,np.sin(x))
ax[1].plot(x,np.cos(x));
```
