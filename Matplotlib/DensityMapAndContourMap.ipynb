```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
import numpy as np
```

```python
#三維函數的視覺化(plt.contour())
def f(x,y):
  return np.sin(x)**10+np.cos(10+y*x)*np.cos(x)
x=np.linspace(0,5,50)
y=np.linspace(0,5,40)
X,Y=np.meshgrid(x,y)#建立二維網格
Z=f(X,Y)
plt.subplot(2,1,1)
plt.contour(X,Y,Z,colors='black');#建立等高線圖
plt.subplot(2,1,2)
plt.contour(X,Y,Z,20,cmap='RdGy');#cmap參數指定一個色彩對應表
```

```python
#實心等高線圖(plt.contourf())
plt.contourf(X,Y,Z,20,cmap='RdGy')
plt.colorbar();
```

```python
#更有效率的方式(plt.imshow())
plt.imshow(Z,extent=[0,5,0,5],origin='lower',
      cmap='RdGy')
plt.colorbar();
```

```python
#合併等高線圖和影像圖
contours=plt.contour(X,Y,Z,colors='black')
plt.clabel(contours,inline='True',fontsize=8)
plt.imshow(Z,extent=[0,5,0,5],origin='lower',
      cmap='RdGy',alpha=0.5)
plt.colorbar();
```
