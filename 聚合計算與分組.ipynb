```
#Pandas簡單聚合運算
import numpy as np
import pandas as pd
import seaborn as sns
#Series之聚合計算會回傳單一值
ser=pd.Series(np.random.rand(5))
print(ser)
print(ser.sum())
print(ser.mean())
#DataFrame之聚合計算會以每欄為主
df=pd.DataFrame({'A':np.random.rand(5),
         'B':np.random.rand(5)})
print(df)
print(df.mean())
print(df.mean(axis='columns'))#along the columns做計算
#行星範例
planets=sns.load_dataset('planets')#此資料已經在seaborn內
print(planets)
print(planets.dropna().describe())#對每一欄做各種聚合計算
```

```
#GroupBy:Split,Apply,Combine
import numpy as np
import pandas as pd
import seaborn as sns
#split-apply-combine
df=pd.DataFrame({'key':['A','B','C','A','B','C'],
         'data':range(6)})
print(df)
print(df.groupby('key').sum())
#GroupBy物件
planets=sns.load_dataset('planets')
print(planets[planets['method']=='Eclipse Timing Variations'])
print(planets.groupby('method')['orbital_period'].median())
for method,group in planets.groupby('method'):
  print("{0:30s} shape={1}".format(method,group.shape))
#分派方法
print(planets.groupby('method')['year'].describe())
```

```
#聚合計算、過濾、轉換和套用
import numpy as np
import pandas as pd
df=pd.DataFrame({'key':['A','B','C','A','B','C'],
         'data1':range(6),
         'data2':np.random.randint(0,10,6)})
print(df)
#聚合計算(可用字串或是函式,且可以一次輸入多個)
print(df.groupby('key').aggregate(['min',np.median,max]))
print(df.groupby('key').aggregate({'data1':'min',
                  'data2':'max'}))#data1做min然後data2做max
#過濾
def filter_func(x):
  return x['data2'].std()>4 #為一個過濾的標準
print(filter_func(df))
print(df.groupby('key').std())
print(df.groupby('key').filter(filter_func))
#轉換(輸出形狀與輸入是一樣的)
print(df.groupby('key').transform(lambda x:x-x.mean()))
#apply()
def norm_by_data2(x):
  x['data1']/=x['data2'].sum()
  return x
print(df.groupby('key').apply(norm_by_data2))
```

```
#指定分割
import numpy as np
import pandas as pd
#分組鍵list,陣列,序列數,或索引
L=[0,1,0,1,2,0]
df=pd.DataFrame({'key':['A','B','C','A','B','C'],
         'data1':range(6),
         'data2':np.random.randint(0,10,6)})
print(df)
print(df.groupby(L).sum())
print(df.groupby(df['key']).sum())
#字典對應索引值
df2=df.set_index('key')
mapping={'A':'vowel','B':'consonant','C':'consonant'}
print(df2)
print(df2.groupby(mapping).sum())
#python函式
print(df2.groupby(str.lower).mean())
#有效鍵list
#(先以str.lower groupy再以mapping group)
print(df2.groupby([str.lower,mapping]).mean())
#(先以mapping groupy再以str.lower group)
print(df2.groupby([mapping,str.lower]).maen())
```

```
#群組範例
import seaborn as sns
import numpy as np
import pandas as pd

planets=sns.load_dataset('planets')
decade=10*(planets['year']//10)
decade=decade.astype(str)+'s'
decade.name='decade'
print(decade)
#先以methid為group,再細分以decade為group,最後再將number值總和加起來
print(planets.groupby(['method',decade])['number'].sum().unstack().fillna(0))
```
