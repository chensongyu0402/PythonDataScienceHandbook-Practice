```
#一個多重索引的Series
#較不好的方式(手刻)
import numpy as np
import pandas as pd

index=[('California',2000),('California',2010),
    ('New York',2000),('New York',2010),
    ('Texas',2000),('Texas',2010)]
populations=[33871648,37253956,
      18976457,19378102,
      20851820,25145561]
popu=pd.Series(populations,index=index)
print(popu)
print(popu[('California',2010):('Texas',2000)])
print(popu[[i for i in popu.index if i[1] == 2000]])#輸出2000年的紀錄
#較好的方法(利用MultiIndex)
index=pd.MultiIndex.from_tuples(index)
print(index)
popu=popu.reindex(index)
print(popu)
print(popu[:,2010])
#多重索引當作額外維度
popu_df=popu.unstack()#將multiIndex Series轉成DataFrame
print(popu_df)
print(popu_df.stack())#反向操作
popu_df=pd.DataFrame({'total':popu,
            'under18':[9267089,9284094,
                  4687374,4318033,
                  5906301,6879014]})
print(popu_df)
#計算2000與2010年18歲以下之比例
f_u18=popu_df['under18']/popu_df['total']
f_u18=f_u18.unstack()
print(f_u18)
```

```
#MultiIndex建立方法
import numpy as np
import pandas as pd
#利用list
df=pd.DataFrame(np.random.rand(4,2),
        index=[['a','a','b','b'],[1,2,1,2]],
        columns=['data1','data2'])
print(df)
#tuple為key之字典,Pandas預設為MultiIndex
data={('California',2000):33871648,
   ('California',2010):37253956,
   ('New York',2000):20851820,
   ('New York',2010):25145561,
   ('Texas',2000):18976457,
   ('Texas',2010):19378102}
dataS=pd.Series(data)
print(dataS)
#明確指定MultiIndex建構子
print(pd.MultiIndex.from_arrays([['a','a','b','b'],[1,2,1,2]]))
print(pd.MultiIndex.from_tuples([('a',1),('a',2),('b',1),('b',2)]))
print(pd.MultiIndex.from_product([['a','b'],[1,2]]))
#多重索引的階層名稱
dataS.index.names=['state','year']
print(dataS)
#欄的MultiIndex
index=pd.MultiIndex.from_product([[2013,2014],[1,2]],
                 names=['year','visit'])
columns=pd.MultiIndex.from_product([['Bob','Guido','Sue'],['HR','Temp']],
                  names=['subject','type'])
data=np.round(np.random.randn(4,6),1)#與底下兩行,建立仿製資料
data[:,::2]*=10
data+=37
health_data=pd.DataFrame(data,index=index,columns=columns)
print(health_data)
print(health_data['Bob'])
```

```
#MultiIndex的索引和切片
import numpy as np
import pandas as pd

populations={('California',2000):33871648,
   ('California',2010):37253956,
   ('New York',2000):20851820,
   ('New York',2010):25145561,
   ('Texas',2000):18976457,
   ('Texas',2010):19378102}
popu=pd.Series(populations)
popu.index.names=['state','year']
print(popu)
#多重索引的Series
print(popu['California',2000])
print(popu['California'])
print(popu.loc['California':'New York'])
print(popu[:,2000])
print(popu>22000000)
print(popu[popu>22000000])
print(popu[['California','Texas']])
#多重索引的DataFrame
index=pd.MultiIndex.from_product([[2013,2014],[1,2]],
                names=['year','visit'])
columns=pd.MultiIndex.from_product([['Bob','Guido','Sue'],['HR','Temp']],
                  names=['subject','type'])
data=np.round(np.random.randn(4,6),1)
data[:,::2]*=10
data+=37
health_data=pd.DataFrame(data,index=index,columns=columns)
print(health_data)
print(health_data['Guido','HR'])
print(health_data.loc[:,('Bob','HR')])
print(health_data.iloc[:2,:2])
idx=pd.IndexSlice
print(health_data.loc[idx[:,1],idx[:,'HR']])
```

```
#重排列多重索引
import numpy as np
import pandas as pd

index=pd.MultiIndex.from_product([['a','c','b'],[1,2]],
                 names=['char','int'])
data=pd.Series(np.random.rand(6),index=index)
print(data)
#沒有排序會產生錯誤
try:
  data['a':'b']
except KeyError as e:
  print(type(e))
  print(e)
#所以要先排序
data=data.sort_index()
print(data)
print(data['a':'b'])
#索引的stacking和unstacking
populations={('California',2000):33871648,
   ('California',2010):37253956,
   ('New York',2000):20851820,
   ('New York',2010):25145561,
   ('Texas',2000):18976457,
   ('Texas',2010):19378102}
popu=pd.Series(populations)
popu.index.names=['state','year']
print(popu.unstack(level=0))
print(popu.unstack(level=1))
print(popu.unstack().stack())
#索引的設定與重設
popu_flat=popu.reset_index(name='population')
print(popu_flat)
print(popu_flat.set_index(['state','year']))#將原始輸入資料轉成MultiIndex
```

```
#多重索引上資料聚合處理
import numpy as np
import pandas as pd

index=pd.MultiIndex.from_product([[2013,2014],[1,2]],
                 names=['year','visit'])
columns=pd.MultiIndex.from_product([['Bob','Guido','Sue'],['HR','Temp']],
                  names=['subject','type'])
data=np.round(np.random.randn(4,6),1)#與底下兩行,建立仿製資料
data[:,::2]*=10
data+=37
health_data=pd.DataFrame(data,index=index,columns=columns)
print(health_data)
#計算每一年2次測量平均
data_mean=health_data.mean(level='year')
print(data_mean)
#以欄(axis=1)為主階層平均
print(data_mean.mean(axis=1,level='type'))
```
