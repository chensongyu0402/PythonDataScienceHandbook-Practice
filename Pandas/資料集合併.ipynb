```
#串接
import numpy as np
import pandas as pd
#由函示建立DataFrane
def make_df(cols,ind):
  data={c:[str(c)+str(i) for i in ind]
        for c in cols}
  return pd.DataFrame(data,ind)
print(make_df('ABC',range(3)))
#Numpy串接
x=[1,2,3]
y=[4,5,6]
z=[7,8,9]
print(np.concatenate([x,y,z]))
x=[[1,2],
  [3,4]]
print(np.concatenate([x,x],axis=1))
print(np.concatenate([x,x]))
#pd.concat進行簡單串接
ser1=pd.Series(['A','B','C'],index=[1,2,3])
ser2=pd.Series(['D','E','F'],index=[4,5,6])
print(pd.concat([ser1,ser2]))
df1=make_df('AB',[1,2])
df2=make_df('AB',[3,4])
print(df2)
print(pd.concat([df1,df2]))
df3=make_df('AB',[1,2])
df4=make_df('CD',[1,2])
print(pd.concat([df3,df4],axis='columns'))
```

```
#重複索引之解決
import numpy as np
import pandas as pd

def make_df(cols,ind):
  data={c:[str(c)+str(i) for i in ind]
        for c in cols}
  return pd.DataFrame(data,ind)

x=make_df('AB',[0,1])
y=make_df('AB',[2,3])
print(x)
print(y)
print(pd.concat([x,y]))
#產生例外
try:
  pd.concat([x,y],verify_integrity=True)
except ValueError as e:
  print('ValueError:',e)
#忽略索引
print(pd.concat([x,y],ignore_index=True))
#加上多重所引鍵
print(pd.concat([x,y],keys=['x','y']))
```

```
#使用join串接
import numpy as np
import pandas as pd

def make_df(cols,ind):
  data={c:[str(c)+str(i) for i in ind]
        for c in cols }
  return pd.DataFrame(data,ind)

df5=make_df('ABC',[1,2])
df6=make_df('BCD',[3,4])
print(pd.concat([df5,df6]))
print(pd.concat([df5,df6],join='inner'))#inner交集
print(pd.concat([df5,df6],join='outer'))#outer聯集
```
