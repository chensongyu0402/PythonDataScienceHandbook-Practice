```
#正常index
import numpy as np
import pandas as pd
data=pd.Series([0.25,0.5,0.75,1.0])
print(data)
print(data.values)
print(data.index)
print(data[1])
print(data[1:3])
```

```
#指定index
import numpy as np
import pandas as pd
#字串
data=pd.Series([0.25,0.5,0.75,1.0],
        index=['a','b','c','d'])
print(data)
print(data['b'])
#非連續整數
data=pd.Series([0.25,0.5,0.75,1.0],
        index=[2,5,3,7])
print(data)
print(data[3])
```

```
#特殊化字典
import numpy as np
import pandas as pd
population_dict={'California':38332521,
         'Texas':26448193,
         'New York':19651127,
         'Florida':19552860,
         'Illinois':12882135}
population=pd.Series(population_dict)
print(population)
print(population['Texas'])
print(population['California':'New York'])#切片
```

```
#建構Series物件
import numpy as np
import pandas as pd
#index預設
pd1=pd.Series([2,4,6])
print(pd1)
#純量重複填入指定索引
pd2=pd.Series(5,index=[100,200,300])
print(pd2)
#排序index或是values
pd3=pd.Series({2:'a',1:'b',3:'c'})
print(pd3)
print(pd3.sort_index())#對index排序
print(pd3.sort_values())#對values排序
#明確指定index(有明確指出才會output)
pd4=pd.Series({2:'a',1:'b',3:'c'},index=[3,2])
print(pd4)
```

```
#Pandas DataFrame物件
import numpy as np
import pandas as pd
population_dict={'California':38332521,
         'Texas':26448193,
         'New York':19651127,
         'Florida':19552860,
         'Illinois':12882135}
area_dict={'California':423967,
         'Texas':695662,
         'New York':141297,
         'Florida':170312,
         'Illinois':149995}
area=pd.Series(area_dict)
population=pd.Series(population_dict)
print(area)
states=pd.DataFrame({'population':population,
           'area':area})
print(states)
print(states.index)
print(states.columns)
print(states['area'])#特殊化字典
```

```
#建構DataFrame物件
import numpy as np
import pandas as pd
#單行DataFrame
population_dict={'California':38332521,
         'Texas':26448193,
         'New York':19651127,
         'Florida':19552860,
         'Illinois':12882135}
population=pd.Series(population_dict)
DataP=pd.DataFrame(population,columns=['population'])
print(DataP)
#字典的list建立
data=[{'a':i,'b':2*i}for i in range(3)]
print(data)
DataD1=pd.DataFrame(data)
print(DataD1)
#沒有值用NaN替代
DataD2=pd.DataFrame([{'a':1,'b':2},{'a':3,'c':4}])
print(DataD2)
#從NumPy的二微陣列建立
DataD3=pd.DataFrame(np.random.rand(3,2),
          columns=['foo','bar'],
          index=['a','b','c'])
print(DataD3)
#從NumPy結構陣列
A=np.zeros(3,dtype=[('A','i8'),('B','f8')])
print(A)
print(pd.DataFrame(A))
```

```
#與NumPy有許多相同處，但不同於NumPy其index無法修改(例如:ind[0]=1)
import numpy as np
import pandas as pd
ind=pd.Index([2,3,5,7,11])
print(ind)
print(ind[1])
print(ind[::2])
print(ind.size,ind.shape,ind.ndim,ind.dtype)
```

```
#index當作有序集合
import numpy as np
import pandas as pd
indA=pd.Index([1,3,5,7,9])
indB=pd.Index([2,3,5,7,11])
print(indA&indB)
print(indA|indB)
print(indA^indB)#XOR(對稱差)((A-B)or(B-A))
```
