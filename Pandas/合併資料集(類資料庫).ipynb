```
#join類別
#一對一join
import numpy as np
import pandas as pd

df1=pd.DataFrame({'employee':['Bob','Jake','Lisa','Sue'],
         'group':['Accounting','Engineering','Engineering','HR']})
df2=pd.DataFrame({'employee':['Lisa','Bob','Jake','Sue'],
         'hire_date':[2004,2008,2012,2014]})
print(df1)
print(df2)
df3=pd.merge(df1,df2)
print(df3)
#多對一join
df4=pd.DataFrame({'group':['Accounting','Engineering','HR'],
         'supervisor':['Carly','Guido','Steve']})
print(df4)
print(pd.merge(df3,df4))
#多對多join
df5=pd.DataFrame({'group':['Accounting','Accounting','Engineering'
             ,'Engineering','HR','HR'],
         'skills':['math','spreadsheets','coding','linux',
              'spreadsheets','organization']})
print(df5)
print(pd.merge(df1,df5))
```

```
#Merge Key 規格
import numpy as np
import pandas as pd
#On關鍵字
df1=pd.DataFrame({'employee':['Bob','Jake','Lisa','Sue'],
         'group':['Accounting','Engineering','Engineering','HR']})
df2=pd.DataFrame({'employee':['Lisa','Bob','Jake','Sue'],
         'hire_date':[2004,2008,2012,2014]})
print(pd.merge(df1,df2,on='employee'))
#Left_on和Right_on關鍵字
df3=pd.DataFrame({'name':['Bob','Jake','Lisa','Sue'],
         'salary':[70000,80000,120000,90000]})
print(pd.merge(df1,df3,left_on='employee',right_on='name'))
print(pd.merge(df1,df3,left_on='employee',right_on='name').drop('name',axis=1))
print
#Lefr_index和Right_index關鍵字
df1a=df1.set_index('employee')
df2a=df2.set_index('employee')
print(df1a,df2a)
print(pd.merge(df1a,df2a,left_index=True,right_index=True))
#join預設以索引為key
print(df1a.join(df2a))
#混合索引與欄位
print(pd.merge(df1a,df3,left_index=True,right_on='name'))
```

```
#在join指定集合算數運算
import numpy as np
import pandas as pd

df6=pd.DataFrame({'name':['Peter','Paul','Mary'],
         'food':['fish','beans','bread']})
df7=pd.DataFrame({'name':['Mary','Joseph'],
         'drink':['wine','bear']})
print(df6)
print(df7)
#預設為inner join
print(pd.merge(df6,df7))
#利用how指定使用不同方式
print(pd.merge(df6,df7,how='outer'))
#以左邊項目(left join)為主或是右邊項目(right join)為主
print(pd.merge(df6,df7,how='left'))
print(pd.merge(df6,df7,how='right'))
```

```
#重疊欄位名稱(suffixes)
import numpy as np
import pandas as pd
df8=pd.DataFrame({'name':['Bob','Jake','Lisa','Sue'],
         'rank':[1,2,3,4]})
df9=pd.DataFrame({'name':['Bob','Jake','Lisa','Sue'],
         'rank':[3,1,4,2]})
print(df8)
print(df9)
#相衝突的欄位名稱
print(pd.merge(df8,df9,on='name'))
#使用suffixes自訂
print(pd.merge(df8,df9,on='name',suffixes=['L','R']))
```

```
#範例:美國聯邦州的資料
import numpy as np
import pandas as pd
popu=pd.read_csv('/content/drive/MyDrive/資料科學家/DS-Coding/python資料科學學習手冊/data/state-population.csv')
areas=pd.read_csv('/content/drive/MyDrive/資料科學家/DS-Coding/python資料科學學習手冊/data/state-areas.csv')
abbrevs=pd.read_csv('/content/drive/MyDrive/資料科學家/DS-Coding/python資料科學學習手冊/data/state-abbrevs.csv')
#data的imformation
print(popu)
print(areas.head())
print(abbrevs.head())
#先將popu與abbrevs做merge
merged=pd.merge(popu,abbrevs,left_on='state/region',right_on='abbreviation',how='left')
merged=merged.drop('abbreviation',axis=1)
print(merged)
#確認有無空值
print(merged.isnull().any())
print(merged[merged['population'].isnull()])
print(merged.loc[merged['state'].isnull(),'state/region'].unique())
#把state之空值補上
merged.loc[merged['state/region']=='PR','state']='Puerto Rico'
merged.loc[merged['state/region']=='USA','state']='United States'
#再合併一次
final=pd.merge(merged,areas,on='state',how='left')
#再確認有無空值
print(final.isnull().any())
print(final['state'][final['area (sq. mi)'].isnull()].unique())
#選擇丟棄該值(1.因為USA與PR對此資料無影響 2.所以,最後兩者的資料會被丟棄(USA沒有area,PR沒有populations))
final.dropna(inplace=True)
#選擇我們最想要的資料
data2010=final.query("year == 2010 & ages == 'total' ")
print(data2010)
#重新索引(以便最後可以方便閱讀資料),再計算各州密度
data2010.set_index('state',inplace=True)
density=data2010['population']/data2010['area (sq. mi)']
density.sort_values(ascending=False,inplace=True)#降序排列
print(density.head())#知道前5筆data(預設)
print(density.tail())#知道末端data
print(density)
```
