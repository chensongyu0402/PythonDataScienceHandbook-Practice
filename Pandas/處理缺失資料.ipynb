```
#缺失資料None和NaN
import numpy as np
import pandas as pd
#None(無法拿來運算，因為會出錯)
vals1=np.array([1,None,3,4])
print(vals1)
#NaN為IEEE浮點數表達方式(可以拿來運算)
vals2=np.array([1,np.nan,3,4])
print(vals2)
print(vals2.dtype)
print(1+np.nan)
print(0*np.nan)
print(vals2.sum(),vals2.min(),vals2.max())
print(np.nansum(vals2),np.nanmin(vals2),np.nanmax(vals2))#可以排除NaN做計算
```

```
#在Pandas中的NaN與None
import numpy as np
import pandas as pd
x=pd.Series([1,np.nan,2,None])
print(x)
x=pd.Series(range(2),dtype=int)
print(x)
x[0]=None
print(x)
```

```
#在Null值操作
import numpy as np
import pandas as pd
data=pd.Series([1,np.nan,'hello',None])
#偵測空值
print(data.isnull())
print(data[data.notnull()])
#拋棄空值
print(data.dropna())
df=pd.DataFrame([[1,np.nan,2],
        [2,3,5],
        [np.nan,4,6]])
print(df)
print(df.dropna())#移除任何出現null值的那些列
print(df.dropna(axis='columns'))
df[3]=np.nan
print(df)
print(df.dropna(axis='columns',how='all'))#只有那欄全是Null才會移除
print(df.dropna(axis='rows',thresh=3))#只有那列非空值至少3個才會輸出
#填入空值
data=pd.Series([1,np.nan,2,None,3],index=list('abcde'))
print(data)
print(data.fillna(0))#空值填入0
print(data.fillna(method='ffill'))#forward-fill,使用前一個值
print(data.fillna(method='bfill'))#back-fill,使用後一個值
print(df)
print(df.fillna(method='ffill',axis=1))#沿著列forwatd-fill
```
