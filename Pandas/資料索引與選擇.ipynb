```
#Series
#Series當作字典
import numpy as np
import pandas as pd
data=pd.Series([0.25,0.5,0.75,1.0],
        index=['a','b','c','d'])
print(data)
print(data['b'])
print('a' in data)
print(data.keys())
print(list(data.items()))
data['e']=3.14
print(data)
#Series當作一維陣列
print(data['a':'c'])
print(data[0:2])
print(data[(data>0.3)&(data<0.8)])
print(data[['a','e']])
```

```
#indexer:loc,iloc
import numpy as np
import pandas as pd
data=pd.Series(['a','b','c'],index=[1,3,5])
print(data)
#明確索引
print(data.loc[1])
print(data.loc[1:3])
#隱含索引
print(data.iloc[1])
print(data.iloc[1:3])
```

```
#DataFrame中選取資料
#DataFrame作為字典
import numpy as np
import pandas as pd
area=pd.Series({'California':423967,
         'Texas':695662,
         'New York':141297,
         'Florida':170312,
         'Illinois':149995})
population=pd.Series({'California':38332521,
         'Texas':26448193,
         'New York':19651127,
         'Florida':19552860,
         'Illinois':12882135})
data=pd.DataFrame({'area':area,'population':population})
print(data)
print(data['area'])
print(data.area)
print(data.area is data['area'])
data['density']=data['population']/data['area']
print(data)
#DataFrame當作二維陣列
print(data.values)#看成二維陣列
print(data.T)#轉置
print(data.values[0])
print(data.iloc[:3,:2])
print(data.loc[:'Florida',:'population'])
print(data.loc[data.density>100,['population','density']])#遮罩加fancy indexing
data.iloc[0,2]=90 #修改density值
print(data)
#額外索引慣例
print(data['Florida':'Illinois'])
print(data[1:3])
print(data[data.density>100])
```
