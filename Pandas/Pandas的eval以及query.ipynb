```
#複合式操作
import numpy as np
import pandas as pd
import numexpr
#numexpr之使用
x=np.random.rand(int(1E6))
y=np.random.rand(int(1E6))
print(x)
print(y)
mask=(x>0.5)&(y<0.5)
mask_numexpr=numexpr.evaluate('(x>0.5) & (y<0.5)')
#利用np.allclose判斷兩者是否相近
print(np.allclose(mask,mask_numexpr))
```

```
#pd.eval()高效率操作
import numexpr
import numpy as np
import pandas as pd
nrows,ncols=100000,100
df1,df2,df3,df4=(pd.DataFrame(np.random.rand(nrows,ncols))
         for i in range(4))
print(df1.head(),df2.head(),df3.head(),df4.head())
print(pd.eval('df1+df2+df3+df4').head())
#利用np.allclose判斷兩者是否相近
print(np.allclose(df1+df2+df3+df4,pd.eval('df1+df2+df3+df4')))
#支援各種運匴
df1,df2,df3,df4,df5=(pd.DataFrame(np.random.randint(0,1000,(100,3)))
          for i in range(5))
#算術運算子
result1=-df1*df2/(df3+df4)-df5
result2=pd.eval('-df1*df2/(df3+df4)-df5')
print(np.allclose(result1,result2))
#比較運算子
result1=(df1<df2)&(df2<=df3)&(df3!=df4)
result2=pd.eval('(df1<df2)&(df2<=df3)&(df3!=df4)')
print(np.allclose(result1,result2))
#位元運算子
result1=(df1<0.5)&(df2<0.5)|(df3<df4)
result2=pd.eval('(df1<0.5)&(df2<0.5)|(df3<df4)')
print(np.allclose(result1,result2))
#文字形式and以及or
result3=pd.eval('(df1<0.5)and(df2<0.5)or(df3<df4)')
print(np.allclose(result1,result3))
#物件屬性和索引
result1=df2.T[0]+df3.iloc[1]
result2=pd.eval('df2.T[0]+df3.iloc[1]')
print(np.allclose(result1,result2))
```

相同點：計算表達結果

不同點：df.eval若表達式為邏輯，结果返回bool dataFrame;df.query则返回bool dataFrame的實際內容

```
#DataFrame.eval()
import numpy as np
import pandas as pd
df=pd.DataFrame(np.random.rand(1000,3),columns=['A','B','C'])
print(df.head())
#pd.eval
result1=(df['A']+df['B'])/(df['C']-1)
result2=pd.eval('(df.A+df.B)/(df.C-1)')
print(np.allclose(result1,result2))
#DataFrame.eval更簡潔
result3=df.eval('(A+B)/(C-1)')
print(np.allclose(result1,result3))
print(df.head())
#DataFrame.eval()賦值運算
df.eval('D =(A+B)/C',inplace=True)#新的欄位
print(df.head())
df.eval('D =(A-B)/C',inplace=True)#對欄位修改
print(df.head())
#DataFrame.eval()本地端變數
column_mean=df.mean(1)#1:axis=1(每列作mean)
result1=df['A']+column_mean
result2=df.eval('A+@column_mean')
print(np.allclose(result1,result2))
#DataFrame.query()方法
result1=df[(df.A<0.5)&(df.B<0.5)]
result2=df.query('A<0.5 and B<0.5')
print(np.allclose(result1,result2))
#df.query本地端變數
Cmean=df['C'].mean()
result1=df[(df.A<Cmean)&(df.B<Cmean)]
result2=df.query('A<@Cmean and B<@Cmean')
print(np.allclose(result1,result2))
```
