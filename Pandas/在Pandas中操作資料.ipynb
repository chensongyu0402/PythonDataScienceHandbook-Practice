```
#索引保存
import numpy as np
import pandas as pd
ser=pd.Series(np.random.randint(0,10,4))
print(ser)
df=pd.DataFrame(np.random.randint(0,10,(3,4)),columns={'A','B','C','D'})
print(df)
print(np.exp(ser))
print(np.sin(df * np.pi/4))
```

```
#索引對齊
#Series索引對齊
import numpy as np
import pandas as pd
area=pd.Series({'Alasa':1723337,'Texas':695662,'California':423967},name='area')
population=pd.Series({'California':38332521,'Texas':26448193,'New York':19651127}
                     ,name='population')
print(population/area)
print(area.index|population.index)
#資料遺失填入特定值
A=pd.Series([2,4,6],index=[0,1,2])
B=pd.Series([1,3,5],index=[1,2,3])
print(A+B)
print(A.add(B,fill_value=0))
#DataFrame索引對齊
A=pd.DataFrame(np.random.randint(0,20,(2,2)),columns=list('AB'))
print(A)
B=pd.DataFrame(np.random.randint(0,10,(3,3)),columns=list('BAC'))
print(B)
print(A+B)
fill=A.stack().mean()
print(A.add(B,fill_value=fill))
```

```
#DataFrame和Series間操作
import numpy as np
import pandas as pd
A=np.random.randint(10,size=(3,4))
print(A)
print(A-A[0])
df=pd.DataFrame(A,columns=list('QRST'))
print(df-df.iloc[0])
print(df.subtract(df['R'],axis=0))
halfrow=df.iloc[0,::2]
print(halfrow)
print(df-halfrow)
```
