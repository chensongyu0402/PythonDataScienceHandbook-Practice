```
#陣列加總
#一維的時候兩者一樣，但多維時不一樣
import numpy as np
#一維(又以np.sum()較快)
x=np.random.random(100)
print(sum(x))
print(np.sum(x))
print(x.sum())
#多維
x=np.array([[1,2],
      [3,4]])
print(x)
print(sum(x))
print(np.sum(x))
print(x.sum())
```

```
#最大最小值
#np.min()或是np.max()較快
import numpy as np
x=np.random.random(100)
print(min(x))
print(max(x))
print(np.min(x))
print(np.max(x))
```

```
#多維度聚合運算
M=np.random.random((3,4))
print(M)
#陣列元素總和
print(M.sum())
#每行最小或最大值
print(M.min(axis=0))#等價於np.min(M,axis=0)
print(M.max(axis=0))#等價於np.max(M,axis=0)
#每列最大或最小值
print(M.min(axis=1))#等價於np.min(M,axis=1)
print(M.max(axis=1))#等價於np.max(M,axis=1)
```

```
#美國總統平均身高
import numpy as np
import pandas as pd
data=pd.read_csv('/content/drive/MyDrive/資料科學家/練習程式/python資料科學學習手冊/NumPy/data/president_heights.csv')
height=np.array(data['height(cm)'])
print(data)
print(height)
print('Mean height:    ',height.mean())
print('Standard deviation:',height.std())
print('Minimum height:  ',height.min())
print('Maximum height:  ',height.max())
print('25th percentile:',np.percentile(height,25))
print('Median:      ',np.median(height))
print('75th percentile:',np.percentile(height,75))
```

```
#美國總統身高繪圖
import matplotlib.pyplot as plt
import seaborn;seaborn.set()

plt.hist(height)
plt.title('Height Distribution of US Presidents')
plt.xlabel('height(cm)')
plt.ylabel('number')
```
