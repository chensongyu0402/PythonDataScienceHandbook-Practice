```
#陣列運算
import numpy as np
x=np.arange(5)
print(x)
print(x+5)
print(x-5)
print(x*2)
print(x/2)
print(x//2)#取/2得到之商的地板數(floor devide)
print(-x)
print(x**2)
print(x%2)
```

```
#絕對值
import numpy as np
#實數
x=np.array([-1,-2,-3,0,4,7,9])
print(abs(x))
print(np.abs(x))
print(np.absolute(x))
#複數
x=np.array([3-4j,4-3j,2+0j,0+1j])
print(np.abs(x))
```

```
#三角函數
import numpy as np

#從0到3.14....平均分散於3個數
theta=np.linspace(0,np.pi,3)
print(theta)
print(np.sin(theta))
print(np.cos(theta))
print(np.tan(theta))

#反三角
x=np.array([-1,0,1])
print(x)
print(np.arcsin(x))
print(np.arccos(x))
print(np.arctan(x))
```

```
#指數與對數
#指數
x=[1,2,3]
print(x)
print(np.exp(x))#e^x
print(np.exp2(x))#2^x
print(np.power(3,x))#3^x

#對數(沒有講就是取自然對數)
x=[1,2,4,10]
print(x)
print(np.log(x))
print(np.log2(x))
print(np.log10(x))

#維持精準度(expm1與log1p互為逆運算)(保證了x資料的有效性)
x=[0.001,0.01,0.1]
print(np.expm1(x)) #為exp(x)-1
print(np.log1p(x)) #為log(1+x)
```

```
#設定輸入
import numpy as np
#將x元素乘以10直接存入y之記憶體內
x=np.arange(5)
y=np.empty(5)
np.multiply(x,10,out=y)
print(x)
print(y)
#存入指定記憶體位置且間隔數為2
y=np.zeros(10)
np.power(2,x,out=y[::2])
print(y)
```

```
#聚合(reduce只有最後結果,accumulate中間結果)
import numpy as np
x=np.arange(1,6)
print(x)
#計算所有元素總和
y=np.add.reduce(x)
print(y)
#計算所有元素乘積
y=np.multiply.reduce(x)
print(y)
#計算每次的元素和
y=np.add.accumulate(x)
print(y)
#計算每次的元素乘積
y=np.multiply.accumulate(x)
print(y)
```
