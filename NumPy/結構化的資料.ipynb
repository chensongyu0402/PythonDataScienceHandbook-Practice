```
#numpy的結構化陣列
import numpy as np
name=['Alice','Bob','Cathy','Doug']
age=[25,45,37,19]
weight=[55.0,85.5,68.0,61.5]
data=np.zeros(4,dtype={'names':('name','age','weight'),
           'formats':('U10','i4','f8')})
print(data.dtype)
data['name']=name
data['age']=age
data['weight']=weight
print(data)
#取得所有name
print(data['name'])
#取得第1列的資料
print(data[0])
#取得最後一列的name
print(data[-1]['name'])
#布林遮罩
print(data[data['age']<30]['name'])
```

```
#進階複合型態
import numpy as np
tp=np.dtype([('id','i8'),('mat','f8',(3,3))])
x=np.zeros(2,dtype=tp)
print(x[0])
print(x['mat'][0])
```

```
#欄位當屬性
import numpy as np
print(data['age'])
#利用np.recarray
data_rec=data.view(np.recarray)
print(data_rec.age)
```
