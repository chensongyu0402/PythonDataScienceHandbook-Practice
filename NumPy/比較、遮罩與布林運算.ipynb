```
#比較
import numpy as np
#一個陣列
x=np.array([1,2,3,4,5])
print(x)
print(x<3)
print(x>3)
print(x<=3)
print(x>=3)
print(x!=3)
print(x==3)
#兩個陣列
print((2*x)==(x**2))
#任意大小和形狀之陣列均可
rng=np.random.RandomState(0)
x=rng.randint(10,size=(3,4))
print(x)
print(x<6)
```

```
#布林陣列
import numpy as np
x=np.random.randint(10,size=(3,4))
print(x)
#兩者一樣
print(np.count_nonzero(x<6))
print(np.sum(x<6))#因為true為1,而false為0
print(np.sum(x<6,axis=1))#每列小於6之個數
#any存在就好,all必須所有都要符合
print(np.any(x>8))
print(np.any(x<0))
print(np.all(x<10))
print(np.all(x==6))
#np.all()和np.any均有此方法
print(np.all(x<8,axis=1))#是否每列均小於8
```

```
#遮罩運算
import numpy as np
x=np.array([[5,0,3,3],
      [7,9,3,5],
      [2,4,7,6]])
print(x<5)
print(x[x<5])
#建立x<5遮罩
y=(x<5)
print(x[y])
#建立特定範圍遮罩
x=np.array([1,2,3,4,5])
y=(np.arange(5)<4)&(np.arange(5)>0)
print(x)
print(x[y])
```

```
#2014西雅圖雨量(圖and布林運算子and遮罩)
import matplotlib.pyplot as plt
import seaborn;seaborn.set()
import pandas as pd
import numpy as np
rainfall=pd.read_csv('/content/drive/MyDrive/資料科學家/練習程式/python資料科學學習手冊/NumPy/data/Seattle2014.csv')['PRCP'].values
inches=rainfall/254
#繪圖
plt.hist(inches,40)
#布林運算
print('0.5<inches<1的天數:',np.sum((inches>0.5)&(inches<1)))
print('沒有下雨的天數:',np.sum(inches==0))
print('下雨的天數:',np.sum(inches>0))
print('雨量>0.5inches的天數:',np.sum(inches>0.5))
print('雨量<0.2inches的天數:',np.sum((inches>0)&(inches<0.2)))
#遮罩
rainy=(inches>0)#建立有雨量的遮罩
summer=(np.arange(365)<262)&(np.arange(365)>172)#所有夏天遮罩(夏天是第172天開始)
print('2014年下雨天的雨量中位數(inches):    ',np.median(inches[rainy]))
print('2014年夏天降雨量的中位數(inches):    ',np.median(inches[summer]))
print('2014年夏天最大降雨量(inches):      ',np.max(inches[summer]))
print('2014年非夏天的下雨天降雨量中位數(inches):',np.median(inches[rainy&(~summer)]))
```
