```
#探索索引
import numpy as np
x=np.random.randint(100,size=10)
print(x)
#直接要第幾個
print([x[3],x[7],x[2]])
#傳遞一個串列或索引陣列去取得同樣的結果
ind=[3,7,2]
print(x[ind])
ind=np.array([[3,7],
       [4,5]])
print(x[ind])
#多維
x=np.arange(12).reshape((3,4))
print(x)
row=np.array([0,1,2])
col=np.array([2,1,3])
print(x[row,col])
print(x[row[:,np.newaxis],col])
```

```
#索引組合應用
import numpy as np
x=np.arange(12).reshape((3,4))
print(x)
print(x[2,[2,0,1]])
print(x[1:,[2,0,1]])
row=np.array([0,1,2])
mask=np.array([1,0,1,0],dtype=bool)#1為要輸出,0為不要輸出
print(x[row[:,np.newaxis],mask])
```

```
#選擇隨機點
import numpy as np
import matplotlib.pyplot as plt
import seaborn;seaborn.set()
mean=[0,0]
cov=[[1,2],[2,5]]
x=np.random.multivariate_normal(mean,cov,100)
plt.scatter(x[:,0],x[:,1])
indices=np.random.choice(x.shape[0],20,replace=False)
print(indices)
selection=x[indices]
print(selection)
plt.scatter(x[:,0],x[:,1],alpha=0.3)
plt.scatter(selection[:,0],selection[:,1],facecolor='none',s=200);
```

```
#Fancy修改陣列
import numpy as np
x=np.arange(10)
i=np.array([2,1,8,4])
x[i]=99
print(x)
x[i]-=10
print(x)
i=[2,3,3,4,4,4]
x[i]+=1
print(x)
x=np.zeros(10)
np.add.at(x,i,1)
print(x)
```

```
#資料裝箱
import numpy as np
import matplotlib.pyplot as plt
x=np.random.randn(100)
bins=np.linspace(-5,5,20)
counts=np.zeros_like(bins)#建立與bin大小一樣的陣列且內容為0
i=np.searchsorted(bins,x)#計算在x內的值屬於bin區間中的哪個
np.add.at(counts,i,1)#將各區間出現的字數累計加到count
plt.plot(bins,counts,drawstyle='steps');
```
