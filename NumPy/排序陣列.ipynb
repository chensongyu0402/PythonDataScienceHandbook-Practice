```
#np.sort和np.argsort
import numpy as np
#排序
x=np.array([2,1,4,3,5])
print(np.sort(x))
y=np.array([9,8,5,7,6])
y.sort()
print(y)
#排序過元素之索引
x=np.array([2,1,4,3,5])
i=np.argsort(x)
print(i)
print(x[i])
```

```
#沿著列或行排序
x=np.random.randint(0,10,(4,6))
print(x)
print(np.sort(x,axis=0))#對每行
print(np.sort(x,axis=1))#對每列
```

```
#分區Partitioning
import numpy as np
#分區
x=np.array([7,2,3,1,6,5,4])
print(np.partition(x,3))
x=np.random.randint(0,10,(4,6))
print(x)
print(np.partition(x,3,axis=0))#對每行
print(np.partition(x,3,axis=1))#對每列
#分區index
x=np.array([7,2,3,1,6,5,4])
i=np.argpartition(x,3)
print(i)
```

```
#k-近鄰算法
import numpy as np
import matplotlib.pyplot as plt
import seaborn;seaborn.set()
x=np.random.rand(10,2)
#dist_sq=np.sum((x[:,np.newaxis,:]-x[np.newaxis,:,:])**2,axis=-1)
difference=x[:,np.newaxis,:]-x[np.newaxis,:,:]
sq_difference=difference**2
dist_sq=sq_difference.sum(-1)

k=2
nearst_partition=np.argpartition(dist_sq,k+1,axis=1)#k+1因為包含自己
plt.scatter(x[:,0],x[:,1],s=100)#建立點圖

k=2
for i in range(x.shape[0]):
  for j in nearst_partition[i,:k+1]:
    plt.plot(*zip(x[j],x[i]),color='black')#建立連線
```
